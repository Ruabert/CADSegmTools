vis mask and image

```python
mask_root = os.path.join(data_root_fp, '2D_DATA/2D_DATA_BULID_1/MASK/{}/out/'.format(n))
mat_root = os.path.join(data_root_fp, '2D_DATA/2D_DATA_BULID_1/IMAGE/{}/out/'.format(n))
TMP_root = os.path.join(data_root_fp, '2D_DATA/2D_DATA_BULID_1/IMAGE/{}/TMP/'.format(n))

two_d_vis_root = os.path.join(data_root_fp, '2D_DATA/2D_DATA_BULID_1/IMAGE/{}'.format(n))

plt.figure(figsize=[100, 100])

mask_fp_list = os.listdir(mask_root)
mat_fp_list = os.listdir(mat_root)
TMP_fp_list = os.listdir(TMP_root)

try:
    mask_fp_list.sort(key=lambda x: int(x.split('.')[0].split('_')[-1]))
    mat_fp_list.sort(key=lambda x: int(x.split('.')[0].split('_')[-1]))
    TMP_fp_list.sort(key=lambda x: int(x.split('.')[0].split('_')[-2]))
except:
    mask_fp_list.sort(key=lambda x: int(x.split('.')[0].split('_')[-2]))
    mat_fp_list.sort(key=lambda x: int(x.split('.')[0].split('_')[-2]))
    TMP_fp_list.sort(key=lambda x: int(x.split('.')[0].split('_')[-3]))

print(mask_fp_list)
print(mat_fp_list)
print(TMP_fp_list)

assert len(mask_fp_list) == len(mat_fp_list)

for i in range(len(mask_fp_list)):

    mat = scio.loadmat(os.path.join(mat_root, mat_fp_list[i]))
    mask = scio.loadmat(os.path.join(mask_root, mask_fp_list[i]))
    TMP_variable = scio.loadmat(os.path.join(TMP_root, TMP_fp_list[i]))


    img = mat["RV"]
    mask = mask["RV_mask"]
    cp = TMP_variable["centerP"][0]
    c_x = cp[0]
    c_y = cp[1]
    print(c_x,c_y)

    img = img.astype(np.uint8)
    mask =mask.astype(np.uint8)
    img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)


    mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)

    mask_254 = np.where(mask == [254, 254, 254], [55, 200, 5], 0)
    mask_1 = np.where(mask == [1, 1, 1],  [255, 10, 45], 0)

    mask = mask_1 + mask_254
    mask[int(c_x), int(c_y), :] = [0,0,255]

    mask = mask.astype(np.uint8)

    img_add = cv2.addWeighted(img, 0.7, mask, 0.3, gamma=0.0)

    plt.subplot(15, 15, i+1)
    plt.imshow(img_add)

plt.savefig(os.path.join(two_d_vis_root, '{}_{}_with_color.png'.format(n, label)))
plt.show()
```
